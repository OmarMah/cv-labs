Although gradcheck is reliable softmax loss, it is possible that for SVM loss, once in a while, a dimension in the gradcheck will not match exactly. What could such a discrepancy be caused by? Note that SVM loss for a sample $(x_i, y_i)$ is defined as: $$L_i = \sum_{j\ne y_i}\max(0, s_j - s_{y_i} + \Delta)$$ where $j$ iterates over all classes except the correct class $y_i$ and $s_j$ denotes the classifier score for $j^{th}$ class. $\Delta$ is a scalar margin. For more information, refer to 'Multiclass Support Vector Machine loss' on [this](https://cs231n.github.io/linear-classify/) page.

*Hint: the SVM loss function is not strictly speaking differentiable.*


$\color{blue}{\textit Your Answer:}$ This looks like a complicated question so let's start by breaking each term down first:

1- The gradcheck compares the analytical calculas gradient and compares it to the numerical gradient estimated by small finite differences.

2- SVM loss as defined above says that the score of the correct class should be higher than the score assigned to every wrong class by at least a margin = delta.

3- The max function is non-differentiable at exactly x=0 as it acts as a corner.

4- The discrepancy is caused when the finite difference step size h used in the numerical gradient calculation crosses the corner. If the test point is at some value x very close to the corner (such that x - h < 0 and x + h > 0), the numerical gradient will calculate a sloped line connecting the two points, resulting in a fractional value between 0 and 1. However, the analytic gradient calculates the subgradient precisely at point x, which will strictly be 0 (if x < 0) or 1 (if x > 0). This causes the numerical and analytical gradients to be different.

5- No, it is not a reason for concern. These are rare isolated point in the domain. During actual training and optimization, the probability of the loss landing on the corner is very small. even at those non differentiable points gradient descent remains highly effective for optimizing the network.

6- A simple one-dimensional example is f(x)=max(0,x).
At x=-0.001 & h=0.01 the function has a corner:
the analytic gradient = 0 as x<0 and the numerical gradient = 0.009/2*0.01 = 0.45 which is clearly different so the gradcheck fails.

7- I think changing the margin delta shifts the location of the non-differentiable point without altering the actual probability of encountering a gradcheck mismatch though i imagine that our score gaps are usually near 0 as this common when using random init so a smaller margin would make these failures more frequent as the corner sits closer to where the scores already are.
A larger margin may make them less frequent at initialization, because the kink is farther away.
In general it would depend on where the distribution of score differences lies.

```python
# Use the validation set to tune hyperparameters (regularization strength and
# learning rate). You should experiment with different ranges for the learning
# rates and regularization strengths; if you are careful you should be able to
# get a classification accuracy of about 0.365 (> 0.36) on the validation set.

# Note: you may see runtime/overflow warnings during hyper-parameter search.
# This may be caused by extreme values, and is not a bug.

# results is dictionary mapping tuples of the form
# (learning_rate, regularization_strength) to tuples of the form
# (training_accuracy, validation_accuracy). The accuracy is simply the fraction
# of data points that are correctly classified.
results = {}
best_val = -1   # The highest validation accuracy that we have seen so far.
best_softmax = None # The Softmax object that achieved the highest validation rate.


# Provided as a reference. You may or may not want to change these hyperparameters
learning_rates = [1e-6,1e-7,1e-8,2e-6, 5e-6]
regularization_strengths = [2.5e4, 1e4, 1e3, 1e2,3e2, 3e3]


for lr in learning_rates:
    for reg in regularization_strengths:
        softmax = Softmax()
        softmax.train(
            X_train,
            y_train,
            learning_rate=lr,
            reg=reg,
            num_iters=1500,
            batch_size=200,
            verbose=False
        )

        y_train_pred = softmax.predict(X_train)
        train_accuracy = np.mean(y_train == y_train_pred)

        y_val_pred = softmax.predict(X_val)
        val_accuracy = np.mean(y_val == y_val_pred)

        results[(lr, reg)] = (train_accuracy, val_accuracy)

        if val_accuracy > best_val:
            best_val = val_accuracy
            best_softmax = softmax



# Print out results.
for lr, reg in sorted(results):
    train_accuracy, val_accuracy = results[(lr, reg)]
    print('lr %e reg %e train accuracy: %f val accuracy: %f' % (
                lr, reg, train_accuracy, val_accuracy))

print('best validation accuracy achieved during cross-validation: %f' % best_val)
```

**Inline Question 4** - *True or False*

Suppose the overall training loss is defined as the sum of the per-datapoint loss over all training examples. It is possible to add a new datapoint to a training set that would change the softmax loss, but leave the SVM loss unchanged.

$\color{blue}{\textit Your Answer:}$ True


$\color{blue}{\textit Your Explanation:}$ The SVM loss for a given datapoint is calculated using the loss function we have shown above. 

If the new datapoint is already classified correctly by the current model and the score of the correct class exceeds the scores of all other wrong classes by at least the margin delta, the loss for that specific datapoint evaluates to exactly 0. Because the overall loss is a sum of per-datapoint losses, adding this point adds 0 to the total sum and doesn't change it.

When looking at the Softmax loss which is calculated using the cross-entropy loss, it continuously squashes the scores into a probability distribution between 0 and 1. The loss will always be strictly greater than 0 because the probability of the correct class can never be exactly 1 unless the correct class score is positive infinity. Therefore, adding any new datapoint will always contribute a small positive value to the total sum, changing the overall Softmax loss.
